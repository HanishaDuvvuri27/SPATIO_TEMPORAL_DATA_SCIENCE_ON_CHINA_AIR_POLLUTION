**EDA FOR AMBIENT AIR QUALITY DATASET**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_air_quality = pd.read_stata("datasets/pollution/panel20200321.dta")  
df_air_quality.info()
```

```python
df_air_quality.head()
```

```python
df_air_quality.describe()
```

```python
df_air_quality.isnull().sum()
```

```python
missing_percentage = df_air_quality.isnull().sum() / len(df_air_quality) * 100
print(missing_percentage) # <10% can be dropped
```

```python
df_air_quality['date'] = pd.to_datetime(df_air_quality['date'], format='%Y%m%d')
df_air_quality = df_air_quality.sort_values(by=['stationcode', 'date'])

df_air_quality['data'] = df_air_quality.groupby('pollutant')['data'].transform(lambda x: x.interpolate(method='linear'))
```

```python
df_air_quality.isnull().sum()
```

```python
sns.histplot(df_air_quality['data'], bins=100, kde=True)
plt.title('Distribution of Air Quality Index (AQI)')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

```python
df_air_quality['edate'] = pd.to_datetime(df_air_quality['edate'])
df_air_quality.groupby('edate')['data'].mean().plot(figsize=(12, 5))
plt.title('Average AQI Over Time')
plt.xlabel('Date')
plt.ylabel('Average AQI')
plt.show()
```

EDA FOR POLLUTION DATASET 

```python
df_p = pd.read_csv('cleaned_cems.csv')
```

```python
# OUTLIERS DETECTION
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_p['concentration'])
plt.title("Boxplot of Concentration Levels")
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x="provorg", y="nfirms_x", data=df_p)
plt.xticks(rotation=90)
plt.title("Firm Numbers Across Regions")
plt.show()
```

```python
#Skewness & Bimodal Distributions
from scipy.stats import skew

skewness_value = skew(df_p['concentration'].dropna())

sns.histplot(df_p['concentration'], kde=True, bins=30)
plt.title(f"Distribution of Concentration (Skewness: {skewness_value:.2f})")
plt.xlabel("Concentration Levels")
plt.ylabel("Frequency")
plt.show()

```

```python
#   Time Series 

df_p['date'] = pd.to_datetime(df_p['date'])

plt.figure(figsize=(14, 6))
sns.lineplot(x='date', y='concentration', data=df_p)
plt.title("Seasonal Trends in Pollution Levels")
plt.xlabel("Date")
plt.ylabel("Concentration")
plt.show()
```

```python
#CORRELATION
plt.figure(figsize=(10, 6))
sns.heatmap(df_p.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Features")
plt.show()
```

```python
#Pollution Concentration 

plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_p['nfirms_x'], y=df_p['concentration'])
plt.title("Pollution vs. Number of Firms")
plt.xlabel("Number of Firms")
plt.ylabel("Pollution Concentration")
plt.show()
```

```python
#Pollution across Regions
import numpy as np
plt.figure(figsize=(14, 6))
sns.barplot(x='provorg', y='concentration', data=df_p, estimator=np.mean)
plt.xticks(rotation=90)
plt.title("Average Pollution Levels Across Regions")
plt.xlabel("Region")
plt.ylabel("Average Concentration")
plt.show()
```
