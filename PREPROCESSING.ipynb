**PREPROCESSING FOR AMBIENT AIR QUALITY DATASET**

```python
import pandas as pd
import numpy as np
from scipy.spatial import cKDTree
```

```python
pollution = pd.read_stata("datasets/pollution/panel20200321.dta")
weather = pd.read_stata("datasets/weather/provmon.dta")
```

```python
pollution.columns = pollution.columns.str.lower().str.replace(" ", "_")
```

```python
pollution["date"] = pd.to_datetime(pollution["date"], format="%Y%m%d")
```

```python
pollution.drop(columns=["edate", "time", "year", "month", "day"], inplace=True)
```

```python
print(pollution.isnull().sum())
```

```python
pollution["data"].fillna(method="ffill", inplace=True)
# Fill missing pollution data using forward-fill (ffill) since pollution data is time-dependent.
```

```python
pollution_daily = pollution.groupby(["date", "provorg"])["data"].mean().reset_index()
pollution_daily.rename(columns={"data": "avg_aqi"}, inplace=True)
```

**Weather Preprocessing**

```python
weather.columns = weather.columns.str.lower().str.replace(" ", "_")
```

```python
weather["date"] = pd.to_datetime(weather["year"].astype(str) + "-" + weather["month"].astype(str) + "-01")
```

```python
weather.drop(columns=["year", "month"], inplace=True)
```

```python
print(weather.isnull().sum())
```

```python
weather.interpolate(method="linear", inplace=True)
# Interpolate missing values for smooth temperature, wind speed, and visibility trends.
```

```python
weather_monthly = weather.groupby(["date", "provorg"]).mean().reset_index()
```

```python
df_merged = pollution_daily.merge(weather_monthly, on=["date", "provorg"], how="left")
```

```python
df_merged.sort_values(by=["provorg", "date"], inplace=True)  # Ensure data is ordered
df_merged[["temp", "wdsp", "prcp", "visib"]] = df_merged.groupby("provorg")[["temp", "wdsp", "prcp", "visib"]].ffill()
```

```python
df_merged[["temp", "wdsp", "prcp", "visib"]] = df_merged.groupby("provorg")[["temp", "wdsp", "prcp", "visib"]].bfill()
```

```python
print(df_merged.isnull().sum())
```

```python
df_merged.to_csv("cleaned_pollution_weather.csv", index=False)
```

**Data Preprocessing for CEMS Dataset**

```python
import pandas as pd

# Load datasets
df_allair = pd.read_stata("datasets/CEMS/allair_firmLNY_balanced.dta")
df_nox = pd.read_stata("datasets/CEMS/NOXnfirms.dta")
df_so2 = pd.read_stata("datasets/CEMS/SO2nfirms.dta")
```

```python
print(df_allair.isnull().sum())
print(df_nox.isnull().sum())
print(df_so2.isnull().sum())
```

```python
reference_date = pd.Timestamp("1960-01-01")

# Convert `edate` to actual date
df_allair["date"] = reference_date + pd.to_timedelta(df_allair["edate"], unit="D")
df_nox["date"] = reference_date + pd.to_timedelta(df_nox["edate"], unit="D")
df_so2["date"] = reference_date + pd.to_timedelta(df_so2["edate"], unit="D")


```

```python
# Drop old `edate` column
df_allair.drop(columns=["edate"], inplace=True)
df_nox.drop(columns=["edate"], inplace=True)
df_so2.drop(columns=["edate"], inplace=True)
```

```python
df_allair.head()
```

```python
df_nox.head()
```

```python
df_so2.head()
```

```python
df_allair['concentration'].fillna(df_allair['concentration'].median(), inplace=True)
df_nox['nfirms'].fillna(df_nox['nfirms'].median(), inplace=True)
df_so2['nfirms'].fillna(df_so2['nfirms'].median(), inplace=True)

df_allair['pollutant'].fillna("Unknown", inplace=True)
df_nox['pollutant'].fillna("Unknown", inplace=True)
df_so2['pollutant'].fillna("Unknown", inplace=True)
```

```python
print(df_allair.dtypes)
print(df_nox.dtypes)
print(df_so2.dtypes)
```

```python
df_allair['concentration'] = df_allair['concentration'].astype(float)
df_nox['nfirms'] = df_nox['nfirms'].astype(int)
df_so2['nfirms'] = df_so2['nfirms'].astype(int)
```

```python
print(df_allair.dtypes)
print(df_nox.dtypes)
print(df_so2.dtypes)
```

```python
print(df_allair.duplicated().sum())
print(df_nox.duplicated().sum())
print(df_so2.duplicated().sum())
```

```python
df_allair.columns = df_allair.columns.str.lower().str.replace(" ", "_")
df_nox.columns = df_nox.columns.str.lower().str.replace(" ", "_")
df_so2.columns = df_so2.columns.str.lower().str.replace(" ", "_")
```

```python
from scipy import stats
import numpy as np
# Compute Z-scores
df_allair = df_allair[(np.abs(stats.zscore(df_allair['concentration'])) < 3)]
df_nox = df_nox[(np.abs(stats.zscore(df_nox['nfirms'])) < 3)]
df_so2 = df_so2[(np.abs(stats.zscore(df_so2['nfirms'])) < 3)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_allair['concentration'] = scaler.fit_transform(df_allair[['concentration']])
df_nox['nfirms'] = scaler.fit_transform(df_nox[['nfirms']])
df_so2['nfirms'] = scaler.fit_transform(df_so2[['nfirms']])
```

```python
df_allair.head()

```

```python
df_combined = df_allair.merge(df_nox, on=["date", "provorg", "dow", "proven"], how="left")\
                       .merge(df_so2, on=["date", "provorg", "dow", "proven"], how="left")

# Save the final dataset
df_combined.to_csv("cleaned_cems.csv", index=False)

df_combined.head()
```
